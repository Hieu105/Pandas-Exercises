### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2 & 3. Create a data dictionary and Assign it to a variable called pokemon

```python
df_csv = pd.read_csv('pokemon.csv')
targets = ['Bulbasaur', 'Charmander', 'Squirtle', 'Caterpie']
pokemon = df_csv[df_csv['Name'].isin(targets)].copy()

pokemon = pokemon.rename(columns={'Name': 'name', 'Type 1': 'type', 'HP': 'hp'})
pokemon['evolution'] = ['Ivysaur', 'Charmeleon', 'Wartortle', 'Metapod']
pokemon['pokedex'] = ['yes', 'no', 'yes', 'no']
pokemon = pokemon.reset_index(drop=True)

print(pokemon[['evolution', 'hp', 'name', 'pokedex', 'type']])
```

### Step 4. Place the order of the columns as name, type, hp, evolution, pokedex

```python
pokemon = pokemon[['name', 'type', 'hp', 'evolution', 'pokedex']]
print(pokemon)
```

### Step 5. Add another column called place, and insert what you have in mind.

```python
pokemon['place'] = ['Pallet Town', 'Cinnabar Island', 'Cerulean City', 'Viridian Forest']
print(pokemon)
```

### Step 6. Present the type of each column

```python
print(pokemon.dtypes)
```
