### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/US_Crime_Rates_1960_2014.csv).

### Step 3. Assign it to a variable called crime.

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/US_Crime_Rates_1960_2014.csv'
crime = pd.read_csv(url)
crime.head()
```

### Step 4. What is the type of the columns?

```python
crime.dtypes
```

##### Have you noticed that the type of Year is int64? But pandas has a different type to name Date. Convert the type of Year to datetime64

```python
crime.Year = pd.to_datetime(crime.Year, format='%Y')
crime.dtypes
```

### Step 5. Set the Year column as the index of the dataframe

```python
crime.set_index('Year', drop=True, inplace=True)
crime.head()
```

### Step 6. Delete the Total column

```python
del crime['Total']
crime.head()
```

### Step 7. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
crimes = crime.resample('10AS').sum()
population = crime['Population'].resample('10AS').max()
crimes['Population'] = population
crimes
```
