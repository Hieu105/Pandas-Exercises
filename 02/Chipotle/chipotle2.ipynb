### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')

if chipo['item_price'].dtype == 'object':
    chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex=False).astype(float)
```

### Step 4. How many products cost more than $10.00?

```python
chipo_one_prod = chipo.drop_duplicates(['item_name', 'quantity', 'choice_description'])
chipo_one_prod = chipo_one_prod[chipo_one_prod['quantity'] == 1]
chipo_one_prod[chipo_one_prod['item_price'] > 10]['item_name'].nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity'])
chipo_one_prod = chipo_filtered[chipo_filtered['quantity'] == 1]
price_per_item = chipo_one_prod[['item_name', 'item_price']].sort_values(by='item_price', ascending=False)
price_per_item
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(by='item_price', ascending=False).head(1)['quantity']
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo_salad = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
len(chipo_salad)
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo_soda = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
len(chipo_soda)
```
